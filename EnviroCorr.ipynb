```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
enviro = pd.read_csv("SampleMap.csv", index_col=0)
```

```python
corrvars = ['Region_SKN', 'Temp',
       'Salinity', 'DO', 'Density', 'FLUOR', 'TURB', 'CDOM', 'NITRAT',
       'NITRIT', 'SILCAT', 'PHSPHT', 'NH4', 'Lat', 'Long']
```

```python
enviro["StnDepth"] = enviro["cStation"].str.cat(enviro["Depth"], sep= "")
```

```python
enviro.drop_duplicates(subset = "StnDepth", inplace = True)
```

```python
enviro.Region_SKN = pd.Categorical(enviro.Region_SKN)
enviro.Region_SKN = enviro.Region_SKN.cat.codes
enviro
```

```python
enviroSurf = enviro[enviro["Depth"]=="Surface"]
enviroSurf = enviroSurf[corrvars]
enviroSurf
```

```python
sns.pairplot(enviroSurf)
```

```python
eheat= sns.heatmap(enviroSurf.corr().abs(), annot=True)
fig = eheat.get_figure()
fig.set_size_inches( 10, 8)
```

0.8 is cutoff for multicollinearity

temp is 0.97 correlated with DO

Lat/Long

```python
enviroSCM = enviro[enviro["Depth"]=="SCM"]
SCMcorrvars = ['Region_SKN', 'Temp', 'm',
       'Salinity', 'DO', 'Density', 'FLUOR', 'TURB', 'CDOM', 'NITRAT',
       'NITRIT', 'SILCAT', 'PHSPHT', 'NH4', 'Lat', 'Long']
enviroSCM = enviroSCM[SCMcorrvars]
eheat= sns.heatmap(enviroSCM.corr().abs(), annot=True)
fig = eheat.get_figure()
fig.set_size_inches( 10, 8)
```

0.8 is cutoff for multicollinearity

SCM - no colinnearity 

```python
enviroMid = enviro[enviro["Depth"]=="Mid"]
enviroMid = enviroMid[corrvars]
eheat= sns.heatmap(enviroMid.corr().abs(), annot=True)
fig = eheat.get_figure()
fig.set_size_inches( 10, 8)
```

0.8 is cutoff for multicollinearity
Temp colinn with DO, CDOM, Nitrate, Silicate, Phosphate

```python
enviroBot = enviro[enviro["Depth"]=="Bottom"]
enviroBot = enviroBot[corrvars]
eheat= sns.heatmap(enviroBot.corr().abs(), annot=True)
fig = eheat.get_figure()
fig.set_size_inches( 10, 8)
```
